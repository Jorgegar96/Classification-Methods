```python
columns = np.unique(df1['clase'])
rows = np.unique(df1['sexo'])

df = df1.groupby(['sexo', 'clase']).size().reset_index(name="Counts")
discrete_dict = {}
for clase in np.unique(df['clase']):
    discrete_dict[clase] = []
    for valor in np.unique(df['sexo']):
        mask = np.logical_and(df["sexo"] == valor, df["clase"] == clase) 
        data = df.loc[mask, "Counts"].values[0]
        discrete_dict[clase].append(data) 
    
stats = pd.DataFrame(discrete_dict, index = np.unique(df['sexo']))
stats.index.name = "Sexo"
stats.style.set_caption("Hello World")
stats.to_excel("./Statistics/Discrete_Features.xlsx", sheet_name="Sexo Stats")
stats.head()
```

```python
def categ_stats(feature, df1):
    columns = np.unique(df1['clase'])
    rows = np.unique(df1[feature])

    df = df1.groupby([feature, 'clase']).size().reset_index(name="Counts")
    index = np.unique(df[feature])
    cols = np.unique(df['clase'])
    init_data = np.zeros((len(index), len(cols)), dtype=int)
    stats = pd.DataFrame(init_data, index=index, columns=cols)
    for clase in np.unique(df['clase']):
        for valor in np.unique(df[feature]):
            mask = np.logical_and(df[feature] == valor, df["clase"] == clase) 
            if True in np.array(mask):
                data = df.loc[mask, "Counts"].values[0]
                stats.at[valor, clase] = data
    stats.index.name = feature
    return stats
#stats.style.set_caption("Hello World")
#stats.to_excel("./Statistics/Discrete_Features.xlsx", sheet_name="Sexo Stats")
#stats.head()
```

```python
stat_results = {}
for feature in df2.columns:
    if feature != 'clase':
        stat_results[feature] = categ_stats(feature, df2)
```
